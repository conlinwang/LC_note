```python
import math
import numpy as np
import matplotlib.pyplot as plt


summation = 0
N = 1000000 +1
maxi = [0, 0] 
diff = np.zeros(N)
for i in xrange(N):
    temp = i * math.sqrt(2)
    temp = temp - float(math.floor(temp))
    summation = summation + temp
    diff[i] = summation - float(i)/2
    if(maxi[1] < summation - float(i)/2):
    	maxi[0] = i
    	maxi[1] = summation - float(i)/2

	
print 'N =', N-1, "The max of diff=", maxi

fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(diff)
plt.plot(maxi[0], maxi[1], 'ro')
ax.annotate(
    'max diff='+str(maxi[1]), xy=(maxi[0], maxi[1]), 
    xytext=(maxi[0]+1, maxi[1]+0.2), arrowprops=dict(facecolor='black', shrink=0.05),
)
plt.ylabel('the diff of summation - float(i)/2')
```
