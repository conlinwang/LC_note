### Save the Partial sums

```python
# Save the Partial sums
class NumArray(object):
    def __init__(self, nums):
        self.Partial_Sum = nums
        for i in xrange(1, len(nums)):
            self.Partial_Sum[i] += self.Partial_Sum[i-1]

    def sumRange(self, i, j):
        if i > 0 :
            return self.Partial_Sum[j] - self.Partial_Sum[i-1]
        else: 
            return self.Partial_Sum[j] - 0


# Your NumArray object will be instantiated and called as such:
nums = [-2, 0, 3, -5, 2, -1]
obj = NumArray(nums)
param_1 = obj.sumRange(0,1)
print 'param_1=', param_1
param_1 = obj.sumRange(0,2)
print 'param_1=', param_1
param_1 = obj.sumRange(0,3)
print 'param_1=', param_1
```
